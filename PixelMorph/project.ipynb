```python
# Image to Grayscale ,Threshold Converter using Numpy
!pip install matplotlib
import numpy as np
from matplotlib.image import imread
import matplotlib.pyplot as plt


# Step1 : Load the image
# imread -> converts the image into Numpy array.

# ex:(500,700,3) -> 500 rows,700 columns(width),3 channels(RGB)
image=imread(r"D:\AI-ML\PixelMorph\mine.jpg")
print("image shape",image.shape)

```

```python
# Step 2: Extract RGB channels
# Numpy concepts- Indexing & slicing

R=image[:,:,0]
G=image[:,:,1]
B=image[:,:,2]


# Step 3: Convert to Grayscale
# Numpy concepts- Broadcasting and element-wise operations

grayscale = 0.2989 * R + 0.5870 * G + 0.1140 * B
print("Grayscale shape",grayscale.shape)
```

```python
# step 4: Apply Threshold
# Syntax : np.where(condition,value_if_true,value_if_false)
# np.where(condition,value_if_true,value_if_false)

threshold_value=128
threshold=np.where(grayscale>threshold_value,255,0)


```

```python
#step 5 : Display Results
plt.figure(figsize=(12,4)) # Creates a new figure/canvas to plot your images.
plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(image.astype(np.uint8)) #uint converts poxel values into integers(0-255)

plt.subplot(1,3,2)
plt.title("Grayscale")
plt.imshow(grayscale,cmap="gray") #cmap="gray" tells matplotlib to display img in grayscale


plt.subplot(1,3,3)
plt.title("Threshold")
plt.imshow(threshold,cmap="gray")

plt.tight_layout() # Automatically adjusts spacing b/w subplots so titles & images don't overlap.
plt.show()


```
